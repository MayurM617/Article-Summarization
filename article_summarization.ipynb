```python
# Import the summarization pipeline from Hugging Face
from transformers import pipeline
```

```python
# Load the summarization model
summarizer = pipeline('summarization', model='facebook/bart-large-cnn')

def summarize_article(article_text):
    # Summarize the article
    summary = summarizer(article_text, max_length=180, min_length=40, do_sample=False)
    return summary[0]['summary_text']


```

```python
def format_summary(summary):
    # Split the summary into sentences
    sentences = summary.split('. ')

    # Format the summary with newlines and tabs
    formatted_summary = ""
    for sentence in sentences:
        formatted_summary += f"\n\t{sentence.strip()}."

    return formatted_summary
```

```python
# Get user input for the article
article = input("Please enter the article you want to summarize: ")

# Summarize the article
summary = summarize_article(article)

# Format the summary
formatted_summary = format_summary(summary)

# Display the formatted summary
print("\nFormatted Summary:")
print(formatted_summary)
```
